```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Plotting the Parameters

### Load in the CSV file 

```python
param_results = pd.read_csv('./results/parameters.csv')
```

```python
param_results.head()
```

```python
param_results.tail()
```

```python
# Extract the columns for x and y values
x_values = param_results['PERIOD']
y_values = param_results['VARPCC']

# Define the years from 0 to 2020, increasing by 5 years
years = np.arange(2020, 2430, 5)

# Plot the data using Matplotlib
plt.plot(years, y_values)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Units')
plt.title('Variance Of Per Capita Consumption')

# Show the plot
plt.show()
```

```python
# Extract the columns for x and y values
x_values = param_results['PERIOD']
y_values = param_results['L']

# Define the years from 2020 to 2430, increasing by 5 years
years = np.arange(2020, 2430, 5)

# Plot the data using Matplotlib
plt.plot(years+1, y_values)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Level of population and Labor (Millions)')
plt.title('Populaton in Millions')

# Show the plot
plt.show()
```

```python
# Extract the columns for x and y values
x_values = param_results['PERIOD']
Cprice_values = param_results['CPRICEBASE']

# Define the years from 0 to 2020, increasing by 5 years
years = np.arange(2020, 2430, 5)

# Plot the data using Matplotlib
plt.plot(years, Cprice_values)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Carbon Price (2019$ per tCO2)')
plt.title('Carbon Price')

# Show the plot
plt.show()
```

```python
# Extract the columns for x and y values
x_values = param_results['PERIOD']
backstop_values = param_results['PBACKTIME']

# Define the years from 0 to 2020, increasing by 5 years
years = np.arange(2020, 2430, 5)

# Plot the data using Matplotlib
plt.plot(years+1, backstop_values)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Cost of Backstop Technology (2019$ per tCO2)')
plt.title('Backstop Technology Price')

# Show the plot
plt.show()
```

```python
# Extract the columns for x and y values
x_values = param_results['PERIOD']
sigma_values = param_results['SIGMA']

# Define the years from 0 to 2020, increasing by 5 years
years = np.arange(2020, 2430, 5)

# Plot the data using Matplotlib
plt.plot(years+1, sigma_values)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Sigma(t)')
plt.title('CO2-Emissions Output Ratio')

# Show the plot
plt.show()
```
